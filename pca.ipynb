```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Load dataset
iris = load_iris()
X = iris.data   # features (4D: sepal length, sepal width, petal length, petal width)
y = iris.target # labels (species)
```

```python
# Step 2: Standardize data (important for PCA)
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis', edgecolor='k', s=80)

plt.legend(handles=scatter.legend_elements()[0], labels=list(iris.target_names))
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset (2D)")
plt.show()

# Step 5: Explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_)
```
